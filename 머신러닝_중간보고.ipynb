## 4. 목표

*  교통의 혼잡의 심각성을 평가하고 주중 다른 시간과 다른 요일의 교통 상황을 모니터링하는 것

## 5. 주요 기능 및 특징

*  기능 1 : 교통 계획과 혼잡 관리 및 교통 흐름 분석에 유용
*  기능 2 : 차량 관계를 파악하고 혼잡한 지역을 식별
*  기능 3 : 인프라 개선 제공 도시 이동성을 향상 시키고 지속 가능한 도시를 만듦
*  기능 4 : 시간이나 날짜의 교통 패턴을 연구하는 기능

## 6. 기대 효과

*  교통 상황을 시간대와 날짜별로 분석하고 혼잡도에 따라 효율적인 이동을 가능하게 하는 기대가 예상된다.

## 10. 결론

* 결론 : 교통 데이터를 가지고 항목별로 분석한 뒤에 목표한 기대를 이뤄 교통 상황에 대한 혼잡도를 나타내는 것

## 11. 기존 프로젝트의 문제점 개선

*   기존 프로젝트는 시간의 단위가 너무 촘촘해서 데이터가 방대해 보일 수 있는데 시간이나 날짜의 단위를 적절하게 조절하여서 더 간결하게 만들면서도 교통 상황을 한 눈에 예측하거나 파악할 수 있게 개선할 필요성이 있습니다.

## 12. 실제 환경을 어떻게 개선하고 반영할 수 있는지 제안

*  기존의 데이터들을 배열하고 정리한 다음 효율적으로 환경을 구성하기 위한 시간과 날짜를 정하고 그 조건에 맞는 데이터를 배치해서 환경을 개선하면 될 것 같습니다. 또한 환경을 개선해서 더욱 정확도가 높은 예측이 가능하도록 구성하는 것이 목표입니다.

## 13. 기존 프로젝트의 성능을 개선

*   기존에 참고한 프로젝트의 성능을 향상시키기 위해 이론 수업시간에 배운 특성 공학과 규제를 사용하여 개선하거나 정규화 기법을 사용하는 방식으로 성능을 개선할 계획입니다.
