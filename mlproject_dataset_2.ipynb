```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

traffic_2 = pd.read_csv('/content/drive/MyDrive/TrafficTwoMonth.csv')
traffic_2.head(10)
```

```python
traffic_2.info()
```

```python
traffic_2.groupby('Traffic Situation').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
traffic_2.groupby('Day of the week').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
traffic_2['CarCount'].plot(kind='hist', bins=20, title='CarCount')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
traffic_2["CarCount"].value_counts()
```

```python
traffic_2.describe()
```

```python
colormap = plt.cm.gist_heat
plt.figure(figsize=(12,12))
```

```python
sns.pairplot(traffic_2, hue='CarCount');
plt.show()
```
