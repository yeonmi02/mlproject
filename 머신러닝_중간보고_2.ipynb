## 2. 머신러닝 코드를 찾아본 결과
- 우리가 목표한 것에 따라 머신러닝 코드를 어떤 것을 사용하면 좋을지 의논하고 고민을 해봤지만 잘 떠오르지 않아 챗 지피티에 물어봐 도움을 받았다.

- 지피티가 알려준 코드 :

from prophet import Prophet

df_prophet = hourly_df.reset_index()[['timestamp', 'traffic_volume']]
df_prophet.columns = ['ds', 'y']

model = Prophet()
model.fit(df_prophet)

future = model.make_future_dataframe(periods=24, freq='H')
forecast = model.predict(future)

model.plot(forecast)
