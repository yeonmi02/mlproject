```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

traffic_1 = pd.read_csv('/content/drive/MyDrive/Traffic.csv')
traffic_1.head(10)
```

```python
traffic_1.groupby('Traffic Situation').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
traffic_1.groupby('Day of the week').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
traffic_1['CarCount'].plot(kind='hist', bins=20, title='CarCount')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
